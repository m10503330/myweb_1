```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def horizontal_data(dataname,stop,end):

    df=pd.read_csv(dataname)
    df=df.drop(0).dropna(axis=1)
    df["Y"]=df["Y"].astype(float)
    df["Y"]=(df["Y"]*100).astype(int).astype(float)/100
    df["X"]=df["X"].astype(float)
    df["X"]=(df["X"]*100).astype(int).astype(float)/100
    df=df.set_index('Y',drop=1)
    indexlist=df.index.unique()

    plt.style.use('ggplot')
    fig,ax=plt.subplots()
    data_type=['U','V','absV']
    style=[':ok',':sr','v:g']
    for i in np.arange(stop,end):
        plt.plot(df[df.index==indexlist[int(i)]]["X"],df[df.index==indexlist[int(i)]]["U"],style[0],lw=1,label=data_type[0])
        plt.plot(df[df.index==indexlist[int(i)]]["X"],df[df.index==indexlist[int(i)]]["V"],style[1],lw=1,label=data_type[1])
        plt.plot(df[df.index==indexlist[int(i)]]["X"],df[df.index==indexlist[int(i)]]["absV"],style[2],lw=1,label=data_type[2])

    ax.legend(title="velocity type")
    ax.xaxis.set_major_locator(plt.MaxNLocator(7))
    ax.set_xlabel('X/D')
    ax.set_ylabel('velocity (m/ms)')
    plt.title("Vertical axis data in Y/D = "+str(indexlist[int(i)]))
    ax.tick_params(colors="black")
    plt.show()
for i in range(50):
    horizontal_data("oil.csv",i,i+1)
    

```
