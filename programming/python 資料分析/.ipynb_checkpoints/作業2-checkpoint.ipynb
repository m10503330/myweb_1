```python
import numpy as np
```

```python
import re
energy=pd.read_excel("https://storage.googleapis.com/py_ml_datasets/Energy%20Indicators.xls")
energy.head();
energy=energy.dropna(thresh=4)
energy=energy.dropna(axis=1)
energy.columns=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
energy=energy.drop(7)
energy[energy[:]=='...']=np.NaN
energy["Energy Supply"]=energy["Energy Supply"]*1000000

dict_data={"Republic of Korea": "South Korea",
"United States of America": "United States",
"United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
"China, Hong Kong Special Administrative Region": "Hong Kong"}

for i in energy["Country"].values:
  energy["Country"][energy["Country"].values==i]=i.split(" (")[0]

for i in energy["Country"].values:
  energy["Country"][energy["Country"].values==i]=re.compile(r'[^\D]+').sub('', i)

for i in dict_data.items():
  energy["Country"][energy["Country"].values==i[0]]=i[1]
```

```python
GDP=pd.read_csv('https://storage.googleapis.com/py_ml_datasets/world_bank.csv')
GDP=GDP.drop([0,1,2])
dict_data={"Korea, Rep.": "South Korea",
"Iran, Islamic Rep.": "Iran",
"Hong Kong SAR, China": "Hong Kong"}
for i in dict_data.items():
  GDP["Data Source"][GDP["Data Source"].values==i[0]]=i[1]
```

```python
ScimEn=pd.read_excel("https://storage.googleapis.com/py_ml_datasets/scimagojr-3.xlsx")
```

```python
ScimEn=ScimEn[ScimEn['Rank']<=15]
GDP2=pd.concat([GDP['Data Source'],GDP.loc[:,'Unnamed: 50':'Unnamed: 59']],axis=1)
GDP2.columns=['Country Name', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
GDP2=GDP2.drop(3)
energy.head()
```

```python
def answer_one(): 
  df=pd.merge(ScimEn,GDP2,left_on='Country',right_on='Country Name')
  df=pd.merge(df,energy,left_on='Country',right_on='Country')
  df.drop("Country Name",axis=1,inplace=True)
  df=df.set_index('Country')
  return df

answer_one()
```
