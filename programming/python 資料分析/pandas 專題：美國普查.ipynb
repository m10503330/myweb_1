# 問題三
哪個郡（county）在 2010-2015 期間人口改變數量的絕對值最高？（考慮 POPESTIMATE2010 到 POPESTIMATE2015 這六個變數） 提示：如果五年的人口數分別為 100, 120, 80, 105, 100, 130 則人口改變數量的絕對值為 |130-80| = 50

````python
df=census_df[census_df['SUMLEV']==50]
maxnum=df[['POPESTIMATE2010','POPESTIMATE2011','POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']].max(axis=1)
minnum=df[['POPESTIMATE2010','POPESTIMATE2011','POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']].min(axis=1)
df['popdef']=maxnum-minnum

def answer_three(df):
  return df.loc[df['popdef'].idxmax()]["CTYNAME"]
answer_three(df)
````
