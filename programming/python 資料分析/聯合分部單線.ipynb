```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
def new_data(data):
    df=pd.read_csv(data+'.csv')
    df=df.drop(0).dropna(axis=1)
    df["Y"]=df["Y"].astype(float)
    df["Y"]=(df["Y"]*100).astype(int).astype(float)/100
    df["X"]=df["X"].astype(float)
    df["X"]=(df["X"]*100).astype(int).astype(float)/100
    df=df.set_index("Y")
    loc_index=df.index.unique()
    df=df[df.index==loc_index[18]]
    print(loc_index[18])
    return df;
df1=new_data("solid")
df2=new_data("air")
df3=new_data("oil")
sns.jointplot("X","absV",data=df1,kind="reg")
sns.jointplot("X","absV",data=df2,kind="reg")
sns.jointplot("X","absV",data=df3,kind="reg")
```
